```python
import pandas as pd#data manipulation
import numpy as np#numerical cal. e.g., matrix
import matplotlib.pyplot as plt#visulization
import seaborn as sns#advanced lib of matplot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python

af = pd.read_csv('appliance_energy.csv')
af.head()
```

```python
#missing values
af.isnull().sum()
```

```python
# feature and target-dependent?
# features-independent
x=af [['Temperature (°C)']]
y= af['Energy Consumption (kWh)']
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,train_size=0.8,random_state=42
)
```

```python
#instantiate mode
model = LinearRegression()
```

```python
#train
model.fit(x_train,y_train
          )
```

```python
#perdiction
y_pred =model.predict(x_test)
y_pred
```

```python
#metrics mse& r2
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2=r2_score(y_test,y_pred)
print(f' R Square Score:{r2*100:.3f}%')
```

```python
plt.scatter(x_test,y_test,color='black',label='test data')
plt.plot(x_test,y_pred,color='gray',label="regression line")
plt.xlabel("Temperature (°C)")
plt.ylabel('Energy Consuption (kmh)')
plt.legend()
plt.title("Energy consuption predection based on linear regression")
plt.show()
```

```python
import joblib
joblib.dump(model,'slrmodel.pkl')
```
